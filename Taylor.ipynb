```python
import matplotlib.pyplot as plt
import numpy as np
#balíčky
```

```python
def ataylor(b,x, r,h=0.1):
        f=np.sin(b)
        d=np.cos(b)
        g=-np.sin(b)
        if r == 1:
            taylor= f + d*(x-b)
        else:
            taylor = f + d*(x-b) +(g/2*(x-b)**2)
        return taylor
    
def ntaylor(b,x,r, h=0.1,):
    f=np.sin(b)
    d = np.sin(b + h)
    p = np.sin(b - h)
    g=(d - f)/h
    l = (p - 2*f + d)/h**2
    if r == 1:
        taylor= f + g*(x-b)
    else:
        taylor= f + g*(x-b) +(l/2*(x-b)**2)
    return taylor
#funkce
```

```python
r = int(input('Zadej řád derivace: '))
#vstupy
```

```python
x = np.linspace(-3, 4*np.pi, 100)
y=np.sin(x)
plt.plot(x,y)

b = [0,2,4,6]
for i in range(len(b)) :
    y = ataylor(b[i],x,r,h=0.1)

    plt.ylim(-5,5) 
    plt.plot(x,y)
    
b = [0,2,4,6]
for i in range(len(b)) :
    y = ntaylor(b[i],x,r,h=0.1)

    plt.ylim(-5,5) 
    plt.plot(x,y, linestyle='--')


plt.title('Graf fce sinus její a aproximace V BODĚ b')
plt.show()
```
